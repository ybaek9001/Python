```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import urllib.request
import os

print("--------------------------------------------------------------------------------------")
print('    ZESPA Co.,Ltd.\n')
print('    백용희대리\n')
print('    제스파 웹사이트에 있는 썸네일을 가져오는 프로그램입니다.\n')
print('    오늘도 힘차고 즐거운 하루 되세요 ^^')
print("--------------------------------------------------------------------------------------")

print("\n원하는 작업의 번호를 입력하세요. \n\n 1.건강기기 제품 이미지 가져오기 \n 2.미용기기 제품 이미지 가져오기 \n 3.생활가전 제품 이미지 가져오기 \n 4.반려용품 이미지 가져오기 \n")
print('***선택한 제품군 이름으로 폴더가 생성되고 이미지 파일명은 제품 모델명으로 저장됩니다.***\n\n\n\n\n')

key= ''
save_folder = ''

while key == '':
    choice = input("번호: ")
    if choice == '1':
        key = '024'
        save_folder = './건강기기 제품/'
        try:
            os.mkdir('./건강기기 제품')
        except FileExistsError :
            print("!!!!!!   선택하신 '건강기기 제품' 폴더가 이미 존재하므로 진행할 수 없습니다.   !!!!!!")
            sys.exit()                      
    
    elif choice == '2':
        key = '031'
        save_folder = './미용기기 제품/'
        try:
            os.mkdir('./미용기기 제품')
        except FileExistsError :
            print("!!!!!!   선택하신 '미용기기 제품' 폴더가 이미 존재하므로 진행할 수 없습니다.    !!!!!!")
            sys.exit()
    
    elif choice == '3':
        key = '033'
        save_folder = './생활가전 제품/'
        try:
            os.mkdir('./생활가전 제품')
        except FileExistsError :
            print("!!!!!!   선택하신 '생활가전 제품' 폴더가 이미 존재하므로 진행할 수 없습니다.  !!!!!!")
            sys.exit()

    elif choice == '4':
        key = '032'
        save_folder = './반려용품/'
        try:
            os.mkdir('./반려용품')
        except FileExistsError :
            print("!!!!!!   선택하신 '반려용품' 폴더가 이미 존재하므로 진행할 수 없습니다.   !!!!!!")
            sys.exit()
   
    else:
        key = ''
        print('잘못 입력했습니다. 다시 번호를 입력하세요.')




for page in range(20):
    page = page + 1
    base_url = 'http://zespa.co.kr/goods/goods_list.php?page='
    end_url = '&cateCd=%s'%key
    site_url = base_url + str(page) + end_url
    html = urlopen(site_url)
    soup = bs(html, "html.parser")
    imgs = soup.find_all(class_='middle gd_image_lazy')
    temp_names = soup.find_all(class_='item_name')
    img_url = []
    model_num = []

    for img in imgs:
        img_url.append(img['data-original'])


    temp = []
    for i in temp_names:
        t = i.get_text()
        t = t.replace('/','_')
        t = t.replace('~','_')
        c = t.count('-')
        if c == 2:
            model_num.append(t.split('-')[-2])
        elif c == 0:
            model_num.append(t)
        elif c == 3 or 4 :
            t = t.split(' ')[-1]
            t = t.strip('-')
            model_num.append(t)

        else:
            temp = t.split('-')[-3:-1]
            merged = temp[0]+temp[1]
            model_num.append(merged)

    for i in range(len(img_url)):
        urllib.request.urlretrieve(img_url[i], save_folder + model_num[i] +'.jpg')

print("선택한 제품군 이름으로 폴더를 생성하고 이미지를 넣었습니다!")


```
